```python
import re
from dataclasses import dataclass
```

```python
@dataclass
class Trabajo:
    id: int
    nombre: str
    proyecto: str
    horas: int
    pago_horas: float
    salario_semanal: float = 0.0
    
```

```python
data_set = [
 "ID:001; NAME: alberto_EINSTEIN ; PROYECTO:Relatividad; HORAS:40; PAGO_H:35.5",
 "id-002, name- marie_CURIE, proyecto-Radiactividad, horas-35, pago_h-40.0",
 "ID:003; NAME:isaac_NEWTON; PROYECTO:Leyes del Movimiento; HORAS:20,PAGO_H:30.0",
 "id:004 - name: nikola_TESLA - proyecto:Corriente Alterna - horas:50 - pago_h:25.0",
 "ID:005; NAME:ada_LOVELACE; PROYECTO:Algoritmos; HORAS:45; PAGO_H:45.0",
 "id-006, name- charles_DARWIN, proyecto-Evolución, horas-30, pago_h-20.0",
 "ID:007; NAME: rosalind_FRANKLIN ; PROYECTO:Estructura ADN; HORAS:38,PAGO_H:32.5",
 "id:008 - name: stephen_HAWKING - proyecto:Agujeros Negros - horas:15 - pago_h:50.0",
 "ID:009; NAME:katherine_JOHNSON; PROYECTO:Trayectorias NASA; HORAS:42,PAGO_H:40.0",
 "id-010, name- grace_HOPPER, proyecto-Compiladores, horas-48, pago_h-38.0"
]
```

```python
lista_de_pagos = []
```

```python

ids_vistos = set()

# 2. Procesamos el data_set
for item in data_set:
    texto_limpio = re.sub(r'(?i)(name|id|proyecto|horas|pago_h)[:\s\-\;]+', ',', item)
    partes = [p.strip(' ,;-') for p in texto_limpio.split(',')]
    partes = [p for p in partes if p]
    
    # Extraemos el ID numérico
    id_val = int(re.search(r'\d+', partes[0]).group())
    
    # 3. Filtramos duplicados
    if id_val not in ids_vistos:
        nombre_val = partes[1].replace("_", " ").title()
        proyecto_val = partes[2].capitalize()
        horas_val = int(partes[3])
        pago_val = float(partes[4])
        
        # Guardamos en la lista
        registro = Trabajo(id_val, nombre_val, proyecto_val, horas_val, pago_val)
        lista_de_pagos.append(registro)
        
        # Marcamos como visto
        ids_vistos.add(id_val)

```

```python
print(f"{'ID':<5}|{'Nombres':<25}|{'Proyecto':<25}|{'Horas':<15}|{'Pago por Hora':<15}")


# Bucle para imprimir cada registro
for tra in lista_de_pagos:
    # Aquí es donde aplicamos :03 para el ID
    print(f"{tra.id:03}|{tra.nombre:<25}|{tra.proyecto:<25}|{tra.horas:<15}|{tra.pago_horas:<15}") 
```

Parte 2 

```python
print (f"{'ID':<5}| {'Proyecto':<25}| {'Salario semanal':<15}")


for tra in lista_de_pagos:
    salario_semanal= tra.horas * tra.pago_horas 
    print(f"{tra.id:03} | {tra.proyecto:<25} | {salario_semanal:<15}") 
```

```python
salario_alto = 0
investigador = None

for info in lista_de_pagos :
    salario= info.horas * info.pago_horas 
    if salario > salario_alto :
        salario_alto = salario 
        investigador = info

print ( f"El investigador con mas salario es {investigador.nombre} con un salario de {salario} ")      
```

```python
total_horas_trabajadas = 0 

for trabajo in lista_de_pagos :
    total_horas_trabajadas = trabajo.horas + total_horas_trabajadas
    promedio_horas_trabajadas = total_horas_trabajadas / len(lista_de_pagos) 

print (f"El promedio de las horas trabajadas es : {promedio_horas_trabajadas}") 
```
