```python
# Lista de canciones
canciones = []

# Ingreso de 3 canciones
for i in range(3):
    cancion = input(f"Ingrese la canción {i+1}: ")
    canciones.append(cancion)

# Mostrar canciones
print("\nLista de canciones:")
for c in canciones:
    print(c)

# Preguntar si desea agregar canción
opcion = input("\n¿Desea agregar otra canción? (si/no): ")
if opcion == "si":
    nueva = input("Ingrese la nueva canción: ")
    canciones.append(nueva)

# Preguntar si desea eliminar canción
opcion = input("\n¿Desea eliminar una canción? (si/no): ")
if opcion == "si":
    eliminar = input("Ingrese la canción a eliminar: ")
    if eliminar in canciones:
        canciones.remove(eliminar)
        print("Canción eliminada.")
    else:
        print("Canción no encontrada.")

# Mostrar lista actualizada
print("\nLista actualizada:")
for c in canciones:
    print(c)

# Ordenar canciones
canciones.sort()

# Mostrar lista ordenada
print("\nCanciones ordenadas:")
for c in canciones:
    print(c)
```
